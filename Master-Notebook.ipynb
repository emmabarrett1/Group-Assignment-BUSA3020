**Task 2, Question 1**: Clean **all** numerical features and the target variable `price` so that they can be used in training algorithms. For instance, `host_response_rate` feature is in object format containing both numerical values and text. Extract numerical values (or equivalently eliminate the text) so that the numerical values can be used as a regular feature.  
(2 marks)

```python
#keep ID column
df_train_ID = df_train['ID'] 
df_test_ID = df_test['ID']

#drop ID column since it is not necessary for analysis
df_train.drop("ID", axis = 1, inplace = True) 

#drop ID column since it is not necessary for analysis
df_test.drop("ID", axis = 1, inplace = True)

# Concatenate training and test datasets for imputation
df = pd.concat([df_train, df_test]).reset_index(drop=True)


num_col=df.select_dtypes(include=['int64','float64']).columns


cat_col= df.select_dtypes(include=['object']).columns

df['host_response_rate'] = df['host_response_rate'].str.replace('%', '')
df['host_response_rate'] = pd.to_numeric(df['host_response_rate'])

df['price'] = df['price'].str.replace('[\$,]', '', regex = True)
df['price'] = pd.to_numeric(df['price'])

df['host_acceptance_rate'] = df['host_acceptance_rate'].str.replace('%', '')
df['host_acceptance_rate'] = pd.to_numeric(df['host_acceptance_rate'])

df['host_since'] = df['host_since'].str.replace('-', '').astype(int)
df['host_since'] = pd.to_numeric(df['host_since'])

df['first_review'] = df['first_review'].str.replace('-', '').astype(np.float64).astype("Int32")
df['first_review'] = pd.to_numeric(df['first_review'])

df['last_review'] = df['last_review'].str.replace('-', '').astype(np.float64).astype("Int32")
df['last_review'] = pd.to_numeric(df['last_review'])
```

First, we merge train and test dataset into a dataset called "df". Then checked the types of each columns in df. We found some numerical features are in categorical features group as their values are mixed with numbers and text. Therefore, we convert these features to numerical feature as they are supposed to be. The features that were converted are 'price', 'host_response_rate', 'host_since', 'host_acceptance_rate', 'first_review' and 'last_review'. Price feature had the sign'$' which made the feature as categorical. Thus, we removed the dollar sign. Similarly, 'host_response_rate' and 'host_acceptance_rate' had '%' signs respectively. We removed '-' signs of dates from 'first_review' and 'last_review' features. We had to removed the sign in order to convert the features to numerical. 

**Task 2, Question 2** Create at least 4 new features from existing features which contain multiple items of information, e.g. creating `email`,  `phone`, `work_email`, etc. from feature `host_verifications`.  
(2 marks)

```python
from sklearn.preprocessing import LabelEncoder
df['cooking allowed'] = df['room_type'] == 'Entire home/apt'
label_encoder = LabelEncoder()
df['cooking allowed'] = label_encoder.fit_transform(df['cooking allowed'])

df['suitable for family'] = (df['accommodates'] >= 4).astype(bool)
df['suitable for family'] = label_encoder.fit_transform(df['suitable for family'])

df.loc[df['beds'] >= 5, 'bed types'] = 'king'
df.loc[df['beds'] < 5, 'bed types'] = 'queen'
df['bed types'] = label_encoder.fit_transform(df['bed types'])

df.loc[df['review_scores_location'] >= 4.50, 'location rating'] = 'excellect'
df.loc[df['review_scores_location'] < 4.50, 'location rating'] = 'normal'
df['location rating'] = label_encoder.fit_transform(df['location rating'])
```

Four new features were created and added to df according to the instruction. Each new feature is based on existing features from df.
The first new feature is called 'cooking allowed' which is based on 'room type'. There are different values in 'room type' feature and the value in cooking allowed is true if the value in 'room type' is 'Entire home/apt' and else, false.
Second feature is 'suitable for family' which is based on 'accommodates' feature in df that shows the number of people an air bnb can accommodate. As for family, the value is set as 4 and above in order to get true value in 'suitable for family' feature, else false.
Third feature is 'bed types' that is based on 'beds'. There are two values such as king and queen. If the amount of bed is 5 and above, the value in 'bed type' will be 'king' while the rest will be 'queen'.
The last feature is called 'location rating' that is based on 'review scores location' of df. if the review is 4.50 and above, the value in 'location rating' is recored as 'excellent' and for below 4.50, it is recorded as 'normal'.
Finally, all the values in new features are converted to numerical values by using LabelEncoder from sklearn.processing.

#### LASSO Regression

```python
from sklearn.linear_model import Lasso
import warnings
warnings.filterwarnings('ignore')

lasso = Lasso(alpha=0.1)
X_train = df_final_train.loc[:, ~df_final_train.columns.isin(['price'])]
y_train = df_final_train['price']
lasso.fit(X=X_train, y=df_final_train['price'])
y_train_pred = lasso.predict(X=X_train)
mse = mean_squared_error(y_train, y_train_pred)
rmse = math.sqrt(mse)
#print(f'LASSO intercept:\n {lasso.intercept_}')
#print(f'LASSO coefficients:\n {lasso.coef_}')
print(f'LASSO RMSE train: {rmse:.3f}')
```

```python
from sklearn.model_selection import GridSearchCV

params = {'alpha':[1, 0.1, 0.01, 0.001, 0.0001]}    

# step-3: perform grid search
# 3.1 specify model
lasso.fit(X=X_train, y=df_final_train['price'])           

# 3.2 call GridSearchCV()
lasso_tuned = GridSearchCV(estimator = lasso, 
                        param_grid = params, 
                        scoring= 'neg_root_mean_squared_error', 
                        cv = 5, 
                        verbose = 1,
                        return_train_score=True)

lasso_tuned.fit(X=X_train, y=df_final_train['price'])
y_train_pred = lasso_tuned.predict(X=X_train)
mse = mean_squared_error(y_train, y_train_pred)
rmse = math.sqrt(mse)
#print(f'LASSO intercept:\n {lasso.intercept_}')
#print(f'LASSO coefficients:\n {lasso.coef_}')
print(f'Tuned LASSO RMSE train: {rmse:.3f}')
```

The Lasso regression was tuned for the hyperparameter alpha. This showed a reduction in RMSE indicating that when the model's hyperparameter was tuned, it was better at predicting the explanatory variable, airbnb prices in Melbourne. Though it should be noted the reduction in RMSE after the hyperparameter was tuned was very small.

```python
params = [{'max_depth': [3,10,20,30,40,50]}]
# step-3: perform grid search
# 3.1 specify model
forest.fit(X=X_train, y=df_final_train['price'])

# 3.2 call GridSearchCV()
forest_tuned = GridSearchCV(estimator = forest, 
                        param_grid = params, 
                        scoring= 'neg_root_mean_squared_error', 
                        cv = 5, 
                        verbose = 1,
                        return_train_score=True)      

# fit the model
forest_tuned.fit(X=X_train, y=df_final_train['price']) 
y_train_pred = forest_tuned.predict(X_train)
#y_test_pred_ridge = ridge.predict(X_test)
mse = mean_squared_error(y_train, y_train_pred)
rmse = math.sqrt(mse)
print(f'Tuned Forest RMSE train: {rmse:.3f}')
```

The Random Forest regression is a non linear regression model, able to explain relationships that are non linear such as quadratics etc. The model is tuned for the hyperparameter of max depth and this tuned model provides the lowest RMSE of all models, making it the best predictor of airbnb prices in Melbourne with the given variables.
