**Task 2, Question 3**: Impute missing values for all features in both training and test datasets. Hint: make sure you do **not** impute the price in the test dataset.
(3 marks)

```python
import pandas as pd
from sklearn.impute import SimpleImputer

# Assuming you have separate DataFrames for training and test datasets named "df_train" and "df_test"

# Concatenate training and test datasets for imputation
concat_df = pd.concat([df_train, df_test])

# Identify numeric and categorical features
numeric_features = concat_df.select_dtypes(include=['float', 'int']).columns
categorical_features = concat_df.select_dtypes(include=['object']).columns

# Exclude "price" from imputation in the test dataset
numeric_features_test = numeric_features.drop('price')

# Impute missing values for numeric features
numeric_imputer = SimpleImputer(strategy='mean')
concat_df[numeric_features] = numeric_imputer.fit_transform(concat_df[numeric_features])

# Impute missing values for categorical features
categorical_imputer = SimpleImputer(strategy='most_frequent')
concat_df[categorical_features] = categorical_imputer.fit_transform(concat_df[categorical_features])

# Split imputed data back into training and test datasets
df_train_imputed = concat_df[:len(df_train)]
df_test_imputed = df_test.copy()  # Create a copy of the test dataset
df_test_imputed[numeric_features_test] = concat_df[len(df_train):][numeric_features_test]

# Print the imputed DataFrames
print(df_train_imputed)
print(df_test_imputed)
```
